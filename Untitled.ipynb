```python
df.drop(["anime_id"],axis=1,inplace=True)
```

```python
df.columns
x=df
```

```python
#converting in such way that users are features/columns and books are rows
#such that usually for collaborative based recommendation system :
#we need data in such way that: we have user in rows -> labeling diff book or smtg
#so user has identity, and we need their rating
#Then we can convert the data in such way -> pt
#so we can use cosine similarity
#
x
```

```python
pt=x.pivot_table(index='Name',columns='user_id',values='rating')
```

```python
pt_cleaned=pt.fillna(0)
```

```python
#using cosine similarity:

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt_cleaned)
```

```python
pt_cleaned
```

```python
def recommend(name):
    #index_fetch
    index=np.where(pt.index==name)[0][0] #getting index of the element
    """getting the similar items from similarity_scores list, and gettingon accordance
    with similarity score -> thats why key= etc is used ,getting 1-6
"""
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

    for i in similar_items:
        print(pt_cleaned.index[i[0]])
```

```python
recommend("Loveless")
```
